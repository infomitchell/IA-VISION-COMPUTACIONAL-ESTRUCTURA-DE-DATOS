```python
import matplotlib.pyplot as plt
import random    

## aumenta el tamaño de trazado predeterminado
from pylab import rcParams
rcParams['figure.figsize'] = 12,6
```

```python
# crear una funcion para simular el tirado de dados
# los dados imaginarios pueden tomar valores 1 - 100, si el nro. esta entre 1 y 51 la casa gana,
# de 52 a 100 gana el jugador.
def diceroll():
    dice = random.randint(1, 100)
    if dice < 52:
        return(False)
    else:
        return(True)
```

```python
## funcion que simula las apuestas que toma tres argumentos
#1. total_funds = dinero total en mano con el que el jugador comienza
#2. wager_amount = el monto de la apuesta cada vez que el jugador juega
#3. total_plays = la cantidad de veces que el jugador apuesta en este juego

def playy(total_funds, wager_amount, total_plays):
    playnum = []
    funds = []
    ## jugar número 1
    play = 1
    while play <= total_plays:
        ## si el jugador gana
        if diceroll():
            ## agregando el dinero a nuestros fondos
            total_funds = total_funds + wager_amount
            funds.append(total_funds)
            ## agrega el número de juego
            playnum.append(play) 
            
        
        ## si la casa gana
        else:
            playnum.append(play)
            total_funds = total_funds - wager_amount
            funds.append(total_funds)
        play += 1       
    
    ## trama lineal de fondos a lo largo del tiempo    
    plt.plot(playnum, funds)
    finalfund.append(funds[-1])
    return(finalfund)
```

```python
finalfund = []
x = 1

plt.figure(figsize = (12,8))
while x <= 100:
    playy(10000, 100, 1000)
    x += 1
    
plt.xlabel("Numero de jugadas")
plt.ylabel("Valor de la cuenta")
plt.show()

print("el jugador comienza el juego con 10k $ y termina con $", sum(finalfund)/len(finalfund))
```
