```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from PIL import Image
```

```python
image = cv2.imread('Lenna_(test_image).png')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
  
(thresh, blackAndWhiteImage) = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
cv2.imshow('Imagen en blanco y negro', blackAndWhiteImage)
cv2.imshow('Imagen en escala de grises', gray)
cv2.imshow('Imagen original',image) #...imshow usa escala BGR
cv2.waitKey(0)
cv2.destroyAllWindows()

cv2.imwrite('imagen_en_grises.jpg', gray)
cv2.imwrite('imagen_en_byn.jpg', blackAndWhiteImage)
```

```python
#otra forma: usando matplotlib (pyplot)
img = cv2.imread("Lenna_(test_image).png")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)) #plt. usa escala RGB
plt.show()
```
