```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from PIL import Image
```

```python
def change_contrast(img: Image, level: float) -> Image:
    factor = (259 * (level + 255)) / (255 * (259 - level))
    def contrast(c: int) -> float:
        """
        Fundamental Transformation/Operation that'll be performed on
        every bit.
        """
        return 128 + factor * (c - 128)

    return img.point(contrast)
```

```python
with Image.open("Lenna_(test_image).png") as img:
    img= change_contrast(img, 170)
    plt.imshow(img)
    plt.show()
    #img.save("lena_cambio_contraste.png", format="png") #guardar una imagen en el formato que tu desees
```

```python
with Image.open("Lenna_(test_image).png") as img:
    img= change_contrast(img, 170)
    opencvImage = cv2.cvtColor(np.array(img), cv2.COLOR_RGB2BGR)    
    cv2.imshow('Imagen con cambio de contraste',opencvImage)
    cv2.waitKey(0)
    cv2.destroyAllWindows()    
    cv2.imwrite('imagen_con_contraste.jpg', opencvImage)
```
